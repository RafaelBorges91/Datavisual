Aqui selecionarei as principais perguntas de negócio, relacionadas ao escopo "Vendas":

* Qual cidade vendeu mais? :
* Qual representante faz mais vendas? :
* QUal estado e cidade obtiveram maior valor de vendas?:
* Qual os produtos venderam mais e quais clientes faturaram mais:


```python
import pandas as pd
import matplotlib.pyplot as plt

df= pd.read_excel("/content/1 - Base de Dados.xlsx") #Selecionando minha base de dados
display(df)

```

```python
df[['Data_Pedido', 'Nome_Representante', 'Regional', 'Quantidade_Vendida', 'Valor_Total_Venda', 'Nome_Cliente','Cidade_Cliente',	'Estado_Cliente']]
```

```python
# Vamos à primeira pergunta: Qual cidade vendeu mais?

df[['Cidade_Cliente','Quantidade_Vendida']]
```

```python
df[['Cidade_Cliente','Quantidade_Vendida']]
plt.barh(df['Cidade_Cliente'],df['Quantidade_Vendida'], color= 'Turquoise', edgecolor= 'Black') #Vou utilizar a coluna empilhada para melhor visualização
```

```python
# Vamos para a segunda pergunta: Qual representante vendeu mais?

df[['Nome_Representante','Quantidade_Vendida', 'Valor_Total_Venda']]
```

```python
df[['Nome_Representante','Quantidade_Vendida', 'Valor_Total_Venda']]
plt.figure(figsize=(13,5))
plt.bar(df['Nome_Representante'],df['Quantidade_Vendida'], color= 'Turquoise', edgecolor= 'Black')
```

```python
## Qual estado e cidade obtiveram maior valor de vendas?

df[['Cidade_Cliente','Valor_Total_Venda']]
plt.barh(df['Cidade_Cliente'],df['Valor_Total_Venda'], color= 'Turquoise', edgecolor= 'Black')
```

```python
df[['Estado_Cliente','Valor_Total_Venda']]
plt.bar(df['Estado_Cliente'],df['Valor_Total_Venda'], color= 'Turquoise', edgecolor= 'Black')
```

```python
# Qual os produtos venderam mais e quais clientes faturaram mais:

df[['Nome_Produto', 'Quantidade_Vendida', 'Valor_Total_Venda']]
plt.barh(df['Nome_Produto'],df['Quantidade_Vendida'], color= 'Turquoise', edgecolor= 'Black')
```

```python
df[['Nome_Produto', 'Valor_Total_Venda']] #Produto mais vendido: "Notebook Modelo 03"
plt.barh(df['Nome_Produto'],df['Valor_Total_Venda'], color= 'Turquoise', edgecolor= 'Black')
```

```python
df[['Nome_Cliente', 'Valor_Total_Venda']] ## Carrefour é o cliente que mais faturou
plt.figure(figsize=(10,5))
plt.bar(df['Nome_Cliente'],df['Valor_Total_Venda'], color= 'Turquoise', edgecolor= 'Black')
```
